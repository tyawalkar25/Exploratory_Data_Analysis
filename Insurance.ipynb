```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

Insurance = pd.read_csv('insurance_EDA.csv')
Insurance.head()
```

```python
Insurance['charges'].describe()
```

```python
fig, ax = plt.subplots(figsize =(10, 5)) 
ax.hist(Insurance['charges'], bins = 15)
#plt.style.use('ggplot')
#plt.hist(Insurance['charges'], bins = 3000)
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(10, 5)
sns.scatterplot(x = 'age', y = 'charges', data = Insurance)
```

```python
fig = plt.gcf()
fig.set_size_inches(10, 5)
sns.scatterplot(x = 'age', y = 'charges', hue = 'smoker', data = Insurance)
```

```python
fig = plt.gcf()
fig.set_size_inches(10, 5)
sns.scatterplot(x = 'bmi', y = 'charges', data = Insurance)
```

```python
corr_bmi_charges = pearsonr(Insurance['bmi'],Insurance['charges'])[0]
corr_bmi_charges
```

```python
corr_age_charges = pearsonr(Insurance['age'],Insurance['charges'])[0]
corr_age_charges
```
